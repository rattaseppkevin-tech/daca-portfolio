```python
import os
import pandas as pd
from dotenv import load_dotenv
from supabase import create_client
import plotly.express as px

# Laeb .env failist muutujad
load_dotenv()

SUPABASE_URL = os.getenv("SUPABASE_URL")
SUPABASE_KEY = os.getenv("SUPABASE_KEY")

# Luuakse ühendus Supabase-ga
supabase = create_client(SUPABASE_URL, SUPABASE_KEY)

# Päritakse andmed tabelist 'sales'
response = supabase.table('sales').select('*').execute()

# Prinditakse URL kontrolliks ja luuakse DataFrame
df = pd.DataFrame(response.data)

print(df.head())
print(df.info())

df["sale_date"] = pd.to_datetime(df["sale_date"])

daily_sales = df.groupby("sale_date")["total_price"].sum().reset_index()

fig = px.line(daily_sales, x='sale_date', y='total_price', title='Müük ajas')

fig.show()
```

```python
response = supabase.table('sales').select('*').execute()
sales_df = pd.DataFrame(response.data)
response = supabase.table('products').select('*').execute()
products_df = pd.DataFrame(response.data)

merged_df = sales_df.merge(products_df, left_on='product_id', right_on='product_id', suffixes=('_sale', '_product'))

print(merged_df.head())
```
