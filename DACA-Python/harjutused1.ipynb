```python
import pandas as pd
import plotly.express as px
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Loo UrbanStyle näidisandmed (kui CSV puudub)
data = {
    'customer_id': [1001, 1002, 1003, 1001, 1002, 1004, 1003, 1001, 1005, 1004,
                    1002, 1003, 1005, 1001, 1006, 1004, 1002, 1007, 1003, 1005],
    'sale_date': ['2024-01-15', '2024-01-16', '2024-02-01', '2024-02-20', '2024-03-01',
                   '2024-03-05', '2024-03-15', '2024-04-10', '2024-04-12', '2024-04-20',
                   '2024-05-01', '2024-05-10', '2024-05-15', '2024-06-01', '2024-06-05',
                   '2024-06-10', '2024-06-20', '2024-07-01', '2024-07-05', '2024-07-10'],
    'total_price': [89.99, 45.50, 120.00, 67.30, 55.00, 210.00, 33.50, 145.00, 78.00, 92.00,
                     160.00, 44.00, 88.50, 230.00, 37.00, 175.00, 110.00, 65.00, 95.00, 125.00],
    'city': ['Tallinn', 'Tartu', 'Tallinn', 'Tallinn', 'Tartu', 'Pärnu', 'Tallinn', 'Tallinn',
             'Tartu', 'Pärnu', 'Tartu', 'Tallinn', 'Tartu', 'Tallinn', 'Pärnu', 'Pärnu',
             'Tartu', 'Tallinn', 'Tallinn', 'Tartu'],
    'product_category': ['Dresses', 'Tops', 'Denim', 'Accessories', 'Tops', 'Denim', 'Tops',
                         'Dresses', 'Denim', 'Accessories', 'Dresses', 'Tops', 'Denim',
                         'Dresses', 'Accessories', 'Denim', 'Tops', 'Accessories', 'Dresses', 'Denim']
}
df = pd.DataFrame(data)
```

```python
# Uuri andmeid

print("=== SHAPE ===")
print(df.shape)

print("\n=== HEAD ===")
print(df.head())

print("\n=== INFO ===")
print(df.info())
```

```python
print("\n=== DESCRIBE ===")
print(df.describe())

print("\n=== DTYPES ===")
print(df.dtypes)
```

```python
print("Unikaalseid kliente:", df['customer_id'].nunique())
print("Linnad:", df['city'].unique())
print("Tellimused linniti:\n", df['city'].value_counts())
print("Kogukäive:", df['total_price'].sum())
```

```python
# 1. Vaatame, kui palju on unikaalseid tootegruppe
print("Unikaalsete kategooriate arv:", df['product_category'].nunique())

# 2. Vaatame kõigi kategooriate nimesid
print("Kategooriate nimed:", df['product_category'].unique())

# 3. Loeme kokku, mitu ostu igas kategoorias tehti (populaarsus)
print("\nOstude arv kategooriate lõikes:")
df['product_category'].value_counts()
```

```python
# 1. Filtreerimine (WHERE)
print("=== Tallinna tellimused ===")
tallinn = df[df['city'] == 'Tallinn']
print(f"Kokku: {len(tallinn)} tellimust, käive: {tallinn['total_price'].sum():.2f} EUR")

# 2. Grupeerimie (GROUP BY)
print("\n=== Käive linniti ===")
city_revenue = df.groupby('city')['total_price'].agg(['sum', 'mean', 'count'])
city_revenue.columns = ['kogukäive', 'keskmine', 'tellimusi']
print(city_revenue.sort_values('kogukäive', ascending=False))

# 3. Uue veeru loomine (CASE WHEN)
df['order_size'] = df['total_price'].apply(
    lambda x: 'Suur (100+)' if x >= 100 else 'Väike (<100)'
)
print("\n=== Tellimused suuruse järgi ===")
print(df['order_size'].value_counts())
```

```python
customer_summary = df.groupby('customer_id')['total_price'].agg(
    kogukulutus='sum',
    keskmine='mean',
    tellimusi='count'
).reset_index()

# 2. Sorteerime kogukulutuse järgi kahanevalt (suurem eespool)
customer_summary = customer_summary.sort_values('kogukulutus', ascending=False)

# 3. Täidame lüngad: kui x > 200, siis 'VIP', muidu 'Tavaline'
customer_summary['vip_status'] = customer_summary['kogukulutus'].apply(
    lambda x: 'VIP' if x > 200 else 'Tavaline'
)

# Kuvame tulemuse ilma print() funktsioonita, et GitHubis ja ekraanil oleks ilus tabel!
customer_summary.head(10)
```

```python
# Kliendiandmed
customers = pd.DataFrame({
    'customer_id': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    'first_name': ['Jüri', 'Kati', 'Maris', 'Peeter', 'Liina', 'Andres', 'Tiina'],
    'email': ['juri@mail.ee', 'kati@mail.ee', 'maris@mail.ee', 'peeter@mail.ee',
              'liina@mail.ee', 'andres@mail.ee', 'tiina@mail.ee']
})

# Ühenda müügiandmetega
merged = pd.merge(df, customers, on='customer_id', how='left')

# SINU KOOD SIIT ALATES:
# 1. Grupeerime nime järgi ja liidame summad
customer_spending = merged.groupby('first_name')['total_price'].sum().reset_index()

# 2. Võtame 5 suurimat kulutajat
top_5_customers = customer_spending.nlargest(5, 'total_price')

# Kuvame tulemuse
top_5_customers
```

```python
# Käive kategooriate kaupa
cat_revenue = df.groupby('product_category')['total_price'].sum().reset_index()
cat_revenue = cat_revenue.sort_values('total_price', ascending=True)

fig = px.bar(
    cat_revenue,
    x='total_price',
    y='product_category',
    orientation='h',
    title='UrbanStyle: Käive tootekategooriate kaupa',
    text='total_price',
    labels={
        'total_price': 'Käive (EUR)',
        'product_category': 'Kategooria'
    },
    color='product_category'
)
fig.update_layout(showlegend=False)
fig.show()
```

```python
# 1. PARANDUS: Muudame veeru igaks juhuks uuesti kuupäeva tüübiks
df['sale_date'] = pd.to_datetime(df['sale_date'])

# Nüüd .dt töötab ilma veata! Teeme aasta-kuu veeru
df['month'] = df['sale_date'].dt.to_period('M').astype(str)

# 2. Arvutame iga kuu kogukäibe
monthly_revenue = df.groupby('month')['total_price'].sum().reset_index()

# 3. Loo joondiagramm
fig_trend = px.line(
    monthly_revenue, 
    x='month', 
    y='total_price', 
    title='Kuukäibe trend',
    markers=True,
    labels={'month': 'Kuu', 'total_price': 'Käive (€)'}
)
fig_trend.show()
```

```python
fig_pie = px.pie(
    customer_summary, 
    names='vip_status', 
    title='Klientide jaotus VIP-staatuse järgi',
    hole=0.3 # Teeb diagrammi keskelt tühjaks (Donut chart), mis on visuaalselt ilusam
)
fig_pie.show()
```

```python
# 1. Grupeerime RFM andmed segmentide kaupa, et näha klientide arvu ja nende toodud kogukäivet
# 1. Kasutame 'customer_summary' tabelit, et näha VIP vs Tavaliste klientide toodud käivet
story_data = customer_summary.groupby('vip_status').agg(
    kogukaive=('kogukulutus', 'sum'),
    klientide_arv=('customer_id', 'count')
).reset_index()

# Sorteerime, et suurem käive oleks eespool
story_data = story_data.sort_values('kogukaive', ascending=False)

# 2. Loo tulpdiagramm
fig_story = px.bar(
    story_data,
    x='vip_status', # Teljeks on nüüd VIP / Tavaline
    y='kogukaive',
    title='Kogukäive (€) kliendistaatuse lõikes: Kus on meie tegelik raha?',
    labels={'vip_status': 'Kliendi staatus', 'kogukaive': 'Kogukäive kokku (€)'},
    color='vip_status',
    text_auto='.0f', 
    hover_data=['klientide_arv']
)
fig_story.show()
```
